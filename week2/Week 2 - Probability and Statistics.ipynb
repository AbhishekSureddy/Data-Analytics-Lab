### Basic moments of one-dimensional data

```python
import math
import numpy as np
import pandas as pd
from scipy import stats
```

```python
import numpy as np


def mean(arr):
    summup = 0
    for i in range(len(arr)):
        summup += arr[i]
    return(summup/len(arr))

def median(arr):
    res = np.sort(arr)
    if(len(arr)%2==0):
        return((res[len(arr)//2]+res[len(arr)//2-1])/2)
    else :
        return(res[len(arr)//2])
    
def mode(arr):
    unique_values = np.unique(arr)    # storing the unique elements
    count = np.zeros(len(arr))        # this stores the number of times a unique value is repeated
    if (len(unique_values) == len(arr)):
        print("Mode does not exist for this array")
        return(None)
    else:
        for i in range(len(unique_values)):
            for j in range(len(arr)):
                if(unique_values[i] == arr[j]):
                    count[i] +=1
        max_freq = max(count)
        # returning modes
        n_mode = 0
        mode_indices = []
        for i in range(len(unique_values)):
            if(count[i] == max_freq):
                n_mode += 1
                mode_indices.append(i)
        return(unique_values[mode_indices])
    
def variance(arr):
    temp_arr = arr - mean(arr)
    variance = np.dot(temp_arr,temp_arr.T)/(len(arr))
    return(variance)

def standard_deviation(arr):
    return(np.sqrt(variance(arr)))
```

```python
# data1
data1 = np.loadtxt('data1.txt')
data1_mean = mean(data1)
data1_median = median(data1)
data1_mode = mode(data1)
data1_variance = variance(data1)
data1_standard_deviation = standard_deviation(data1)

# data2
data2 = np.loadtxt('data2.txt')
data2_mean = mean(data2)
data2_median = median(data2)
data2_mode = mode(data2)
data2_variance = variance(data2)
data2_standard_deviation = standard_deviation(data2)

# data3
data3 = np.loadtxt('data3.txt')
data3_mean = mean(data3)
data3_median = median(data3)
data3_mode = mode(data3)
data3_variance = variance(data3)
data3_standard_deviation = standard_deviation(data3)

# data4
data4 = np.loadtxt('data4.txt')
data4_mean = mean(data4)
data4_median = median(data4)
data4_mode = mode(data4)
data4_variance = variance(data4)
data4_standard_deviation = standard_deviation(data4)
```

```python
from scipy import stats

CI = 0.95
Mu_hat = mean(data1)
std_dev = 2
n = len(data1)
test_stat = stats.norm.ppf((CI + 1)/2)
lower_bound = Mu_hat - test_stat*std_dev/math.sqrt(n)
upper_bound = Mu_hat + test_stat*std_dev/math.sqrt(n)

# output
print("95 % confidence interval of mean :","( "+str(lower_bound)+", "+str(upper_bound)+" )")
print("upper bound :",upper_bound)
print("lower bound :",lower_bound)
```
