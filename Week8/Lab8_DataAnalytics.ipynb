### Question 1:  
### Max Margin Classifier in 1D
**Let's assume two classes in 1D dimension. Class 1 : 9,10,11,12,13,14; Class -1 : -9,-10,-11,-12,-13,-14<br> <br>a) Plot these points. <br><br>b) Are they linearly separable?<br><br>c) Assume a threshold t. Points to one side of the threshold belong to class -1 and points to the other side of the threshold belong to class 1. Find the range of values of t which classifies the two classes completely. <br><br>d) Let's say y_i is the label and can take values -1 or 1. Max margin classifiers aims to find a supporting hyperplane (a boundary between the two classes) such the distance (or margin) from points belonging to either class is maximised. The points closest to the boundary on either side are termed as support vectors. Maximising the perpendicular distance from the support vectors to line will achieve the goal of max margin classifier. Our objective/cost function for the 1D case is cost = max{min(y_i(t-x_i))} where t is the threhold, y_i is the label and x_i is the data point. Plot the cost as a function of the threshold for the range of values of t found in the (c) part. <br><br>e) For what t is the cost maximum? This is the threshold that optimally classifies the two datasets. Now plot the threshold, and the two classes on a graph. Assign appropriate colours to distinguish the two classes and the threshold.<br><br>f) What are the support vectors?<br><br>g) Comment on the threshold. Is it the mid point of the two support vectors in this linearly separable 1D case?**

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Code here
class1 = np.array([9,10,11,12,13,14])
class2 = -1*np.array([9,10,11,12,13,14])
# part - a
# plotting the data points 
print("part - a")
plt.figure()
plt.scatter(class1,[0]*class1.shape[0],c = 'g',label = "class 1")
plt.scatter(class2,[0]*class2.shape[0],c = 'r',label = "class -1")
plt.legend()
plt.title("visualizing class 1 and class -1")
plt.show()

# part - b
print("part - b\n")
print("Yes, They are linearly seperable.\n")

# part - c
print("part - c\n")
print("Any threshold t, between "+str(np.max(class2))+" and "+str(np.min(class1))+" classifies the two classes completely.\n")

# part - d
print("part - d\n")
# -9 and 9 are the support vectors here
t = np.linspace(-9,9,100) # threshold is from -9 to 9
cost = np.zeros(t.shape)
for i in range(len(t)):
    temp1 = np.min(1*(class1-t[i]))    # min(yi(xi-t))
    temp2 = np.min(-1*(class2-t[i]))
    cost[i] = max(temp1,temp2)         # max(min(yi(xi-t)))

# plotting 
plt.figure()
plt.plot(t,cost)
plt.xlabel("t (threshold)")
plt.ylabel("cost")
plt.title("Cost Vs t")
plt.show()

# part - e
print("part - e\n")
print("The cost is maximum for t = 9 and t = -9, the cost value is = 18")
# plotting
plt.figure()
plt.scatter(class1,[0]*class1.shape[0],c = 'g',label = "class 1")
plt.scatter(class2,[0]*class2.shape[0],c = 'r',label = "class -1")
plt.plot([0,0],[-1,1])
plt.legend()
plt.title("visualizing class 1 and class -1")
plt.show()
print("The blue line is the threshold.\n")

# part - f
print("part - f\n")
print("The support vectors are '-9' and '9'\n")

# part - g
print("part - g\n")
print("The Threshold turns out to be the midpoint('0') of the support vectors in this case")
```

### part - b

Yes, They are linearly seperable.

### part - c

Any threshold t, between -9 and 9 classifies the two classes completely.

### part - e

The cost is maximum for t = 9 and t = -9, the cost value is = 18.

### part - f

The support vectors are '-9' and '9'

### part - g

The Threshold turns out to be the midpoint('0') of the support vectors in this case

```python
# Code here
from sklearn.svm import SVC

class0 = np.array([[-1,-2],[1,1],[1.5,1.5],[-2,-1],[0,0],[1,-.05]])
class1 = np.array([[4.1,5.2],[6,8],[4,4],[7,8],[4.5,6],[5.5,6]])

# visualising
X = np.concatenate((class0,class1))
y = [0]*len(class0)+[1]*len(class1)
y = np.array(y)
clf = SVC(kernel = 'linear')
clf.fit(X,y)
xx, yy = np.meshgrid(np.arange(-3,10,0.1),np.arange(-3,10,0.1))
Z = clf.predict(np.c_[xx.ravel(),yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure()
# for boundary
plt.pcolormesh(xx,yy,Z)
plt.scatter(class0[:,0],class0[:,1],c = 'r', label = "class 0")
plt.scatter(class1[:,0],class1[:,1],c = 'g', label = "class 1")
plt.title(" Decision boundary")
plt.legend()
plt.show()
```

### Question 4  
### Introduction of Kernels
**a) Plot two concentric circles of radii 4 and 8 with origin at 0. Sample 100 points uniformly from each of these circles. (use rcos(theta) and rsin(theta) to find the coordinates of the points lying on the circle. r is the radius of the circle. Vary theta from 3.6 degrees to 360 degrees in intervals of 3.6 degrees.)<br><br>b) Is the data linearly separable? <br><br>c)Since it's not, we will transform them to a different space (with probably different number of dimensions). The transformation (which generally boils down to inner product - something that we call a kernel) will make the data linearly separable in the different space. For each point, compute features as [1, x, x^2,xy,y^2,y]. This transforms the data from 2D to 6D. Now, use sklearn's SVM classifier (linear kernel) to classify the transformed feature space into two classes. <br><br>d)Now apply SVM classifier with i) polynomial kernel with degree 2  ii) rbf kernel and perform the classification. Visualize the decision boundary in the original 2D space in both the cases. Report the accuracies. Which performs best and why?<br>**

```python
# ploting
theta = np.linspace(3.6,360,100)
class0 = np.concatenate((4*np.cos(theta*2*np.pi/360)[:,None],(4*np.sin(theta*2*np.pi/360))[:,None]),axis = 1)
class1 = np.concatenate((8*np.cos(theta*2*np.pi/360)[:,None],(8*np.sin(theta*2*np.pi/360))[:,None]),axis = 1)
# part a
print("part - a")
plt.figure(figsize=(10,10))
plt.scatter(class0[:,0],class0[:,1],c = 'r', label = "class 0")
plt.scatter(class1[:,0],class1[:,1],c = 'g', label = "class 1")
plt.title("Concentric circes")
plt.legend()
plt.show()

# part b
print("part - b")
print("The data is not linearly seperable")

# part c
print("part - c")

X = np.concatenate((class0,class1))
Y = [0]*len(class0)+[1]*len(class1)
Y = np.array(Y)

# transformed data(6d)
X_tr = np.ones((X.shape[0],1))
X_tr = np.concatenate((X_tr,X[:,0][:,None],(X[:,0]**2)[:,None],(X[:,0]*X[:,1])[:,None],(X[:,1]**2)[:,None],X[:,1][:,None]),axis = 1)

# fitting SVM, linear kernel and visualizing
clf = SVC(kernel = 'linear')
clf.fit(X_tr,Y)
xx, yy = np.meshgrid(np.arange(-10,10,0.1),np.arange(-10,10,0.1))
x = xx.ravel()
y = yy.ravel()

x_pr = np.ones((x.shape[0],1))
x_pr = np.concatenate((x_pr,x[:,None],(x**2)[:,None],(x*y)[:,None],(y**2)[:,None],y[:,None]),axis = 1)

Z = clf.predict(np.c_[x_pr])
Z = Z.reshape(xx.shape)
plt.figure(figsize=(10,10))
# for boundary
plt.pcolormesh(xx,yy,Z)
plt.scatter(class0[:,0],class0[:,1],c = 'r', label = "class 0")
plt.scatter(class1[:,0],class1[:,1],c = 'g', label = "class 1")
plt.title(" Decision boundary using linear kernel on [1,x,x^2,xy,y^2,y]")
plt.legend()
plt.show() 

# part - d
# polynomial kernel, degree 2
print("part d")
clf = SVC(kernel = 'poly',degree = 2)
clf.fit(X,Y)
xx, yy = np.meshgrid(np.arange(-10,10,0.1),np.arange(-10,10,0.1))
Z = clf.predict(np.c_[xx.ravel(),yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure(figsize=(10,10))
# for boundary
plt.pcolormesh(xx,yy,Z)
plt.scatter(class0[:,0],class0[:,1],c = 'r', label = "class 0")
plt.scatter(class1[:,0],class1[:,1],c = 'g', label = "class 1")
plt.title(" Decision boundary, using poynomial kernel, degree = 2")
plt.legend()
plt.show()

# rbf kernel
clf = SVC(kernel = 'rbf')
clf.fit(X,Y)
xx, yy = np.meshgrid(np.arange(-10,10,0.1),np.arange(-10,10,0.1))
Z = clf.predict(np.c_[xx.ravel(),yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure(figsize=(10,10))
# for boundary
plt.pcolormesh(xx,yy,Z)
plt.scatter(class0[:,0],class0[:,1],c = 'r', label = "class 0")
plt.scatter(class1[:,0],class1[:,1],c = 'g', label = "class 1")
plt.title(" Decision boundary, using rbfl kernel")
plt.legend()
plt.show()
```

## Observations :
- The accuracies of both the models(polynomial and rbf) are 100%, but polynomial kernel generalizes better compared to rbf kernel

### Question 5 
### Visualising effect of kernels
**Load svm_c0.txt and svm_c1.txt. Create labels as 0 and 1 for these correspondingly. Visualise the two classes and plot decision boundaries obtained by the following kernels and corresponding hyperparameters.<br>(a) Linear kernel with C as 1.0 <br>(b) RBF kernel with gamma as [0.1,1.0,10.0,50.0]** 

```python
# Code here
class0 = np.loadtxt('svm_c0.txt')
class1 = np.loadtxt('svm_c1.txt')

# visualising
X = np.concatenate((class0,class1))
y = [0]*len(class0)+[1]*len(class1)
y = np.array(y)

# part a === linear kernel, C = 1.0
print("part a")
clf = SVC(kernel = 'linear',C = 1.0)
clf.fit(X,y)
xx, yy = np.meshgrid(np.arange(-6,6,0.1),np.arange(-6,6,0.1))
Z = clf.predict(np.c_[xx.ravel(),yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure()
# for boundary
plt.pcolormesh(xx,yy,Z)
plt.scatter(class0[:,0],class0[:,1],c = 'r', label = "class 0")
plt.scatter(class1[:,0],class1[:,1],c = 'g', label = "class 1")
plt.title(" Decision boundary using linear kernel, C = 1.0")
plt.legend()
plt.show()

# part b
print("part b")
gamma_val = [0.1, 1.0, 10, 50.0]

for gamma in gamma_val:
    clf = SVC(kernel = 'rbf',gamma = gamma)
    clf.fit(X,y)
    xx, yy = np.meshgrid(np.arange(-6,6,0.1),np.arange(-6,6,0.1))
    Z = clf.predict(np.c_[xx.ravel(),yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.figure()
    # for boundary
    plt.pcolormesh(xx,yy,Z)
    plt.scatter(class0[:,0],class0[:,1],c = 'r', label = "class 0")
    plt.scatter(class1[:,0],class1[:,1],c = 'g', label = "class 1")
    plt.title(" Decision boundary using rbf kernel, Gamma = "+str(gamma))
    plt.legend()
    plt.show()
```

## Observation :

```python
top5 = np.argpartition(means,-5)[-5:]
top5 = top5[::-1]
print("top 5 mean test indices")
print(top5)
print("Top 5 best cases are :")
print(clf.best_params_)
print("{'C': 100, 'degree': 4, 'kernel': 'poly'}")
print("{'C': 100, 'degree': 3, 'kernel': 'poly'}")
print("{'C': 100, 'degree': 2, 'kernel': 'poly'}")
print("{'C': 100, 'degree': 1, 'kernel': 'poly'}")

# confusion matrix
top_params = [
              {'C': 100, 'degree': 4, 'kernel': 'poly'},
              {'C': 100, 'degree': 3, 'kernel': 'poly'},
             {'C': 100, 'degree': 2, 'kernel': 'poly'},
             {'C': 100, 'degree': 1, 'kernel': 'poly'}]
# for best parameter
clf1 = SVC(100,'rbf',gamma = 0.01)
clf1.fit(X_train,Y_train)
Y_pr = clf1.predict(X_test)
print("f1 score :")
print(f1_score(Y_test,Y_pr,average = 'macro'))
print("Confusion matrix :")
print(confusion_matrix(Y_test,Y_pr))
for params in top_params:
    clf1 = SVC(params['C'],params['kernel'],params['degree'])
    clf1.fit(X_train,Y_train)
    Y_pr = clf1.predict(X_test)
    print("f1 score for "+str(params))
    print(f1_score(Y_test,Y_pr,average = 'macro'))
    print("Confusion matrix :")
    print(confusion_matrix(Y_test,Y_pr))
```

## Significance of hyperparameters
- C is for regularization,A smaller value of C(C=0.1) allows to ignore points close to boundary and increase the margin, while a large value of C(C=1000) a large penalty is assigned to errors and margin errors.  Note that as C decreases, the hyperplanes orientation is also changed to maximize the margin for the rest of the data.
- Gamma is kernel width. As gamma increases it is more likely to overfit.
- degree specifies the polynomial kernel degree to be fit.
- kernel specifies the kernel to be fit.
