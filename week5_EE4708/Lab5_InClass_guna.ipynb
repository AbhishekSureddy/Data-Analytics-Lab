# Classification through Hypothesis Testing (Known Distribution)

**Here, the distribution of the data is already known.**

**Consider 2 one-dimensional Gaussian distributions : (1) mean -1 and variance 1 (2) mean 1 and variance 1.** <br> **a)Plot the 2 distributions on the same graph(100 data points each)** <br> **b)plot the likelihood function and the ratio of the likelihood function (a normalized function)**<br>**c) For the points x= -1.5, -0.5, 0, 0.5, 1.5, hypothesize what class they belong to.** <br> **d) Plot rough decision regions for this task. **<br>**e)For a Bayesian setup with equal probability of the hypothesis, compute the error in the estimate (as a function of x).**


**(5 points)**

**Estimated Time: 20 mins**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import warnings
warnings.filterwarnings("ignore")


def norm_pdf(x, mean, sig):
    return (1/np.sqrt(2*np.pi*sig**2))* (np.exp(-(x - mean)**2/(sig**2 * 2)))
x = np.linspace(-5,5,100)
y1 = norm_pdf(x, -1, 1)
y2 = norm_pdf(x, 1, 1)

print('(a)')
plt.figure(figsize = (6,6))
plt.plot(x, y1,'b', label = "mean(-1) & var(1)")
plt.plot(x, y2, 'r', label = 'mean(1) & var(1)')
plt.title("Gaussian Distribution")
plt.legend()
plt.xlabel('x')
plt.ylabel('pdf')
plt.show()

# (b)
# Ratio of Likelihood Functions
like_ratio_1 = y1/y2

print('(b)')
plt.figure(figsize = (6,6))
plt.plot(x, like_ratio_1,'b')
plt.title("Ratio of likelihood function")
plt.xlabel('x')
plt.ylabel('P(X = x/y = -1)/P(X = x/y = 1)')
plt.show()

# (c)
print('(c)')
print("From the graph :")
print("x = -1.5,-0.5 belongs to distribution 1 where mean is -1")
print("x = 1.5,0.5 belong to distribution 2 where mean is 1")
print("x = 0 can belong to any one of the distributions i.e the decision is inconclusive")


# (d)
print('\n(d)')
plt.figure()
plt.plot([0,0],[0,0.5],'r')
plt.plot(x, y1,'b', label = "mean(-1) & var(1)")
plt.plot(x, y2, 'r', label = 'mean(1) & var(1)')
plt.title("Gaussian Distribution along with decision boundary")
plt.xlabel('x')

print("The points to the left of zero are predicted to come from distribution 1 (mean (-1), Var(1)).")

```

```python
import pandas as pd
from scipy.stats import mode
from sklearn.neighbors import KNeighborsClassifier
train_knn = np.vstack((data1[:500,:], data2[:600,:]))
n_neighbours = [10, 100, 1000]
test = np.array([-0.999, 0.898, 0.8, 0.85, 0.115, 0.104, -0.112])

# (a)
print('(a) ')
dist_mat = np.zeros((len(test), len(train_knn)))
for i in range(len(test)):
    t = train_knn[:,0] - test[i]
    dist_mat[i,:] = t**2

for j in n_neighbours:
    print("No. of nearest neighbours taken are: ", j)
    for k in range(dist_mat.shape[0]):
        ind = np.argpartition(dist_mat[k,:], j)
        l = train_knn[ind[:j], 1]
        print("The class to which {} belongs : class{}".format(test[k], int(mode(l)[0][0])))
    print('\n')

# (b)
plt.figure()
plt.scatter(data1[:500,0],[0]*500,label = 'Class0')
plt.scatter(data2[:600,0],[0]*600,label='Class1')
plt.scatter(test,[0]*len(test),label= 'test_data')
plt.title("Visualizing Test points given wrt train data")
plt.legend()
plt.show()

# (c)
# neigh = KNeighborsClassifier(n_neighbors=3)
# neigh.fit(train_knn[:,0], train_knn[:,1])
```

# Naive Bayes (Assumed Distribution) (20 points)

**Here, the distribution of the data is assumed using some prior information. Naive Bayes works on the principle of Bayes' theorem. It assumes a prior fit for the dataset and then classifies the data into two or more classes. The prior can take the form of a Gaussian prior or any other valid probability distribution like multinomial or bernoulli.**

**Use above given train and test data for this exercise.**

**a)Assume a Gaussian prior. Use Bayes rule and Naive Bayes algorithm to find a threshold which will serve as the decision boundary. Points on one side of the boundary belong to class 0 and points on the other side of the boundary belong to class 1. Report the estimated threshold.<br> b) Use the threshold to classify the test points into one of the two classes. Report the accuracy. <br> c) How is Naive Bayes different from Nearest Neighbours(based on training done and inference computation time)?<br> d) Plot the decision region using matplotlib.
Estimated Time: 40 mins**

```python
train_dat = np.vstack((X_train, Y_train))
train_dat = train_dat.T
tr_0 = train_dat[train_dat[:,1] == 0]
tr_1 = train_dat[train_dat[:,1] == 1]

#(a)
# Assumption ->
# P(X=x|Y=0) ~ N(mu_o, sigma_o)
# P(X=x|Y=1) ~ N(mu_1, sigma_1)

# getting the statistics like mean and standard deviation for each class
s_mean_0 = np.mean(tr_0[:,0])
s_std_0 = np.sqrt(np.var(tr_0))
s_mean_1 = np.mean(tr_1[:,0])
s_std_1 = np.sqrt(np.var(tr_1))

# computing class0 probability in the training dataset
theta = len(tr_0)/(len(tr_1)+len(tr_0))

# Assumed Gaussian Prior
# to get the threshold , we need to solve for the roots of the quadratic equation in x
quad_coef = [1/s_std_1**2-1/s_std_0**2, 2*(s_mean_0/s_std_0**2 - s_mean_1/s_std_1**2), s_mean_1**2/s_std_1**2 - s_mean_0**2/s_std_0**2 + 2*np.log(s_std_1*theta/((1-theta)*s_std_0))]
roots = np.roots(quad_coef)
roots = roots[(roots < max(train_dat[:,0]))]
roots = roots[(roots > min(train_dat[:,0]))]

print("(a)")
print("The Estimated threshold is :", roots)
print("If x < "+ str(roots[0])+", we classify it as Y = 0, else we classify as Y = 1")

#(b)
print('\n(b)')
test_dat = np.vstack((X_test, Y_test))
test_dat = test_dat.T
y_pred = []
for j in range(test_dat.shape[0]):
    if (test_dat[j, 0] > roots[0]):
        y_pred.append(1)
    else:
        y_pred.append(0)
        
y_pred = np.array(y_pred)
print("Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0],(test_dat[:,1] != y_pred).sum()))
accuracy = np.count_nonzero((y_pred - test_dat[:,1]==0))/len(test_dat)*100
print("The accuracy of the model is : "+ str(accuracy)+" %\n")

# plotting the assumed prior distribution of both the classes
x = np.linspace(-5,5,100)
dist_0 = 1/(np.sqrt(2*np.pi)*s_std_0)*np.exp(-1/2*((x-s_mean_0)/s_std_0)**2)
dist_1 = 1/(np.sqrt(2*np.pi)*s_std_1)*np.exp(-1/2*((x-s_mean_1)/s_std_1)**2)

plt.figure(figsize=(10,8))
plt.plot(x,dist_0,'b',label="class0")
plt.plot(x,dist_1,'r',label = "class1")
plt.legend()
plt.title("Gaussian Priors for both classes")
plt.xlabel('x')
plt.ylabel('pdf')
plt.show()

# (c)
print("\n(c)")
print(" 1) Naive Bayes is faster than KNN when it involves larger datasets while both are equally effective on training small and     medium datasets.\n 2) Naive Bayes is faster during classification than KNN \n 3) Naive Bayes improves the classification performance by removing irrelevant features.\n 4) Naive BAyes is not sensitive to local structure of the data whereas KNN is.\n 5) Naive Bayes model is Less accurate than KNN model.  ")

# (d)
print("\n(d)")
plt.figure()
plt.plot([roots[0]]*2,[-1,1])
plt.title("Decision boundary @ x = "+str(roots[0]))
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from matplotlib.colors import ListedColormap

# (a)
# loading the data
X_data = np.loadtxt('LogReg_2d_data.txt')
Y_data = np.loadtxt('LogReg_2d_labels.txt')
X_train, X_test, Y_train, Y_test = train_test_split(X_data,Y_data, test_size = 0.25, random_state = 0)

# logistic regression model
LogReg = LogisticRegression()
LogReg.fit(X_train,Y_train)

# (b)
print("(b)")
# computing accuracy of the model
y_pred = LogReg.predict(X_test)
accuracy = LogReg.score(X_test, Y_test)
print("Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0],(Y_test != y_pred).sum()))
print("Accuracy of the model: ", accuracy * 100, "%")

# computing F1_Score
print("F1 score :",f1_score(Y_test,y_pred))

# (c)
print('\n(c)')
xx, yy = np.meshgrid(np.arange(4,8,0.01),np.arange(1,5,0.01))
Z = LogReg.predict(np.c_[xx.ravel(),yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure()
# for boundary
colors = ('pink', 'cyan')
cmap = ListedColormap(colors[:len(np.unique(Y_test))]) 
plt.pcolormesh(xx,yy,Z, cmap=cmap)
plt.scatter(X_test[Y_test == 0,0],X_test[Y_test == 0,1], label = 'Y = 0')
plt.scatter(X_test[Y_test == 1,0],X_test[Y_test == 1,1], label = 'Y = 1')
plt.title("Decision boundary")
plt.legend()
plt.show()
```
