# Time Series :
    A sequential set of data points measured over time.  
    x(t), t- variable that repersents the elapsed time and t=0,1,2,3,....
### How Time Series is special? 
    It is time dependent. So basic assumption of regression/classification model that the observations are independent doesn't hold in this. 

### Examples:
    Daily temperature
    monthly rain fall
    daily closing price of company stock
    continuous monitoring of a person's heart rate. 

# Auto Regression
    A linear model that uses the dependent relationship between an observation and some number of lagged observations.
    
\begin{equation*}
X_t=\phi_1 X_{t-1}+\phi_2 X_{t-1}....\phi_p X_{t-p}+ w_t \\
X_t= \sum_{i=1}^p \phi_i X_{t-i} + w_t \\
w_t : white noise \sim wn(0,\sigma _w ^2)
\end{equation*}

p : hyper parameter that repersents the length of the direct look back in the series

## Q1) Genarate a time series of 1000 data points with following AR(p) models.
1. AR(0) with $\sigma_w^2=0.3$
2. AR(1) with $\sigma_w^2=0.3 , \phi_1=0.3$
3. AR(1) with $\sigma_w^2=0.3 , \phi_1=0.9$
4. AR(2) with $\sigma_w^2=0.3 , \phi_1=0.4, \phi_2=0.4$
5. AR(2) with $\sigma_w^2=0.3 , \phi_1=0.4, \phi_2=-0.4$


Visualize all 5 time series using subplots and give proper legends and titles. 

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt

# AR(0) with 2=0.3
n = 1000
sigma_sq = 0.3
X1_t = np.sqrt(sigma_sq)*np.random.randn(n,1)

plt.figure(figsize = (15,5))
plt.plot(X1_t)
plt.title("AR(0) with sigma^2w=0.3")
plt.show()

# AR(1) with 2=0.3, 1=0.3
n = 1000
X2_t = np.sqrt(sigma_sq)*np.random.randn(n,1)
phi1 = 0.3
for i in range(1,n):
    X2_t[i] += phi1 * X2_t[i-1]


plt.figure(figsize = (15,5))
plt.plot(X2_t)
plt.title("AR(1) with sigma^2w=0.3, phi1 = 0.3")
plt.show()

# AR(1) with 2=0.3, 1=0.9
n = 1000
X3_t = np.sqrt(sigma_sq)*np.random.randn(n,1)
phi1 = 0.9
for i in range(1,n):
    X3_t[i] += phi1 * X3_t[i-1]


plt.figure(figsize = (15,5))
plt.plot(X3_t)
plt.title("AR(1) with sigma^2w=0.3, phi1 = 0.9")
plt.show()


# AR(2) with 2=0.3, 1=0.4 , phi2 = 0.4
n = 1000
X4_t = np.sqrt(sigma_sq)*np.random.randn(n,1)
phi1 = 0.4
phi2 = 0.4

for i in range(2,n):
    X4_t[i] = X4_t[i] + phi1 * X4_t[i-1] + phi2 * X4_t[i-2]


plt.figure(figsize = (15,5))
plt.plot(X4_t)
plt.title("AR(2) with sigma^2w=0.3, phi1 = 0.4, phi2 = 0.4")
plt.show()

# AR(2) with 2=0.3, 1=0.4 , phi2 = -0.4
n = 1000
X5_t = np.sqrt(sigma_sq)*np.random.randn(n,1)
phi1 = 0.4
phi2 = 0.4

for i in range(2,n):
    X5_t[i] = X5_t[i] + phi1 * X5_t[i-1] + phi2 * X5_t[i-2]


plt.figure(figsize = (15,5))
plt.plot(X5_t)
plt.title("AR(2) with sigma^2w=0.3, phi1 = 0.4, phi2 = -0.4")
plt.show()
```

## Q2)  Genarate a time series of 1000 data points with following AR(p) models.
1. AR(1) with $\sigma_w^2=0.3 , \phi_1=1.005$
2. AR(1) with $\sigma_w^2=0.3 , \phi_1=-1.005$



Visualize the time series using subplots and give proper legends and titles. 
State how it is different from times series generated from Q1.

```python
# 1. AR(1) with 2=0.3, 1=1.005
n = 1000
X1_t = np.sqrt(sigma_sq)*np.random.randn(n,1)
phi11 = 1.005
for i in range(1,n):
    X1_t[i] += phi11 * X1_t[i-1]

# 2. AR(1) with 2=0.3, 1=-1.005
X2_t = np.sqrt(sigma_sq)*np.random.randn(n,1)
phi12 = -1.005
for i in range(1,n):
    X2_t[i] += phi12 * X2_t[i-1]

plt.figure(figsize = (15,10))
plt.subplot(2,1,1)
plt.plot(X1_t, label="phi1 = 1.005")
plt.title("AR(1) with sigma^2w=0.3")
plt.legend()

plt.subplot(2,1,2)
plt.plot(X2_t, label="phi1 = -1.005")
plt.title("AR(1) with sigma^2w=0.3")
plt.legend()
plt.show()
```

## Observation :
- Here, the predictions are increasing exponentially(becoming unstable), when the $ \phi > 1$. Here the effect of past values increase with time.
- where as, when $\phi < 1$ the effect of past values decreases with time.

## Q3) Generate AR(1)  time series of 1000 data points with following parameters and visualize their autocovariance. 
1. AR(1) with $\sigma_w^2=0.3 , \phi_1=0.1$

3. AR(1) with $\sigma_w^2=0.3 , \phi_1=-0.4$

## Auto covariance: 
 
$Cov(X_t, X_{t-k})=E[(X_t - \mu)(X_{t-k} -\mu)]$ <br>

$ Cov(X_t, X_{t-k})=\frac {1}{(N-1)} \sum_{i=0}^{N-k} ( x_i - \mu )( x_{i+k} - \mu )$ <br>

To identify the dominant periods in the time series.


```python
# function for finding covariance
np.random.seed(100)
def covariance(X_t, k, mu):
    temp = 0
    N = len(X_t)
    for i in range(0,N-k):
        temp += (X_t[i] - mu)*(X_t[i+k] - mu)
    return (temp/(N-1))

# 1. AR(1) with 2=0.3, 1 = 0.1
n = 1000
X1_t = np.sqrt(sigma_sq)*np.random.randn(n,1)
phi1 = 0.1
for i in range(1,n):
    X1_t[i] += phi1 * X1_t[i-1]

# covariances
cov = np.zeros(50)
mu = 0           # since error is gaussian with mean = 0
for k in range(50):
    cov[k] = covariance(X1_t, k, mu)

# ploting
plt.figure(figsize=(10,5))
lag = np.arange(50)
plt.scatter(lag,cov)
plt.plot(cov,'g')
plt.plot([0,50],[0,0], 'r')
plt.title("covariance of first 50 lags of AR(1) with phi = 0.1")
plt.show()
    
    
# 2. AR(1) with 2=0.3, 1 = 0.8
X2_t = np.sqrt(sigma_sq)*np.random.randn(n,1)
phi2 = 0.8
for i in range(1,n):
    X2_t[i] += phi2 * X2_t[i-1]

# covariances
cov = np.zeros(50)
mu = 0           # since error is gaussian with mean = 0
for k in range(50):
    cov[k] = covariance(X2_t, k, mu)

# ploting
plt.figure(figsize=(10,5))
lag = np.arange(50)
plt.scatter(lag,cov)
plt.plot(cov,'g')
plt.plot([0,50],[0,0], 'r')
plt.title("covariance of first 50 lags of AR(1) with phi = 0.8")
plt.show()
    
    
# 3. AR(1) with 2=0.3, 1 = -0.4
X3_t = np.sqrt(sigma_sq)*np.random.randn(n,1)
phi3 = -0.4
for i in range(1,n):
    X3_t[i] += phi3 * X3_t[i-1]

# covariances
cov = np.zeros(50)
mu = 0           # since error is gaussian with mean = 0
for k in range(50):
    cov[k] = covariance(X3_t, k, mu)

# ploting
plt.figure(figsize=(10,5))
lag = np.arange(50)
plt.scatter(lag,cov)
plt.plot(cov,'g')
plt.plot([0,50],[0,0], 'r')
plt.title("covariance of first 50 lags of AR(1) with phi = -0.4")
plt.show()
    
```

## Q5) Try to fit and forcast time series for a given real time data set using inbuilt ar library.

Train anf test data :

Inbuilt AR function : from statsmodels.tsa.ar_model import AR

```python
from statsmodels.tsa.ar_model import AR

data = pd.read_csv("Electricity_comsumption_hourly.csv")
# values
X=data['DEOK_MW'].values
train, test = X[1:len(X)-12], X[len(X)-12:]

model = AR(train)
model_fit = model.fit()
print("The test values are :")
print(test)
print()
print("The corresponding predicted values are :")
print(model_fit.predict(start = len(train),end = len(train) + len(test) - 1))

print("\n Observations :\n")
print("The predictions accuracy is decreasing with more number of future steps")
```
